```python
import pandas as pd
```

```python
df = pd.read_csv("IMF_FSI_data_analysis.csv")
df.head(10)
```

```python
# Combining  the rows that are split: keep one row per country-indicator
df = df.groupby(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], as_index=False).first()
df.head(10)
```

```python
df['Indicator Name'].unique()
```

```python
# Melting dataframe from wide format to long format

df_long = df.melt(
    id_vars = ['Country Name','Country Code','Indicator Name','Indicator Code'],
    var_name = 'Period',
    value_name ='FSI_Value'
)
df_long
```

```python
# Removeing rows whre value is NaN
df_long.dropna(subset =['FSI_Value'], inplace = True)
```

```python
# Sorting data
df_long.sort_values(by =['Country Name','Indicator Code','Period'], inplace =True)

# Reseting index
df_long.reset_index(drop = True, inplace = True)
```

```python
 df_long.to_csv("Cleaned_IMF_FSI_for_analysis.csv", index=False)

```
