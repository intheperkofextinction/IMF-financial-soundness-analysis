```python
import pandas as pd
```

```python
df = pd.read_csv(r"Filtered_IMF_FSI.csv")
df.head(5)
```

```python
df.info()
```

```python
# Drop 'Attribute' since its useless for anlaysis
if 'Attribute' in df.columns and df['Attribute'].nunique() == 1:
    df.drop(columns=['Attribute'], inplace=True)
```

```python
# droping columns with no values
df.dropna(axis=1,how= 'all', inplace = True)
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Export cleaned data
output_path = r"Cleaned_IMF_FSI.csv"
df.to_csv(output_path, index=False)
```
