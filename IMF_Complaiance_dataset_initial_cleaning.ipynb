```python
import pandas as pd
```

```python
df = pd.read_csv("FSI_07-05-2025 13-37-20-77_timeSeries\FSI_07-05-2025 13-37-20-77_timeSeries.csv", low_memory = False)
df.head(5)
```

```python
# Filtering Countries
target_countries = ['Brazil','Germany','Japan','France','Unitied kingdom']
df = df[df['Country Name'].isin(target_countries)]
```

```python
indicaters = df['Indicator Name'].unique()
indicaters
```

```python
# Filter Indicators
relevant_indicators = [
    'Financial Soundness Indicators, Core Set, Deposit Takers, Capital Adequacy, Regulatory Tier 1 Capital to Risk-Weighted Assets, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Capital Adequacy, Tier 1 Capital to Assets, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Asset Quality, Non-performing Loans to Total Gross Loans, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Capital Adequacy, Non-performing Loans Net of Provisions to Capital, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Earnings and Profitability, Return on Assets, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Liquidity, Liquid Assets to Short Term Liabilities, Percent',
    'Financial Soundness Indicators, Core Set, Deposit Takers, Sensitivity to Market Risk, Net Open Position in Foreign Exchange to Capital, Percent'
]

df = df[df['Indicator Name'].isin(relevant_indicators)]
```

```python
df.to_csv("Filtered_IMF_FSI.csv", index=False)
```
